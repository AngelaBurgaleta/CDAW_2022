##Entorno

```python
# Creamos el entorno
# Me da un error al poner Taxi-v2 y dice que la versión válida es la v3
env = gym.make("Taxi-v3")
env.render()
```

```python
#Cálculo de filas y columnas necesarias
action_size = env.action_space.n
state_size = env.observation_space.n
print("Columnas: " , action_size)
print("Filas: ", state_size)
```

```python
#Matriz de 6 columnas y 500 filas
qtable = np.zeros((state_size, action_size))
print(qtable)
```

##Q-Learning

El objetivo del Q-learning es aprender una serie de normas que le diga a un agente qué acción tomar bajo qué circunstancias. No requiere un modelo del entorno y puede manejar problemas con transiciones estocásticas y recompensas sin requerir adaptaciones. ​ Q-learning puede identificar una norma de acción-selección óptima para cualquier proceso de decisión, dado un tiempo de exploración infinito y una norma parcialmente aleatoria.

Ref: https://es.wikipedia.org/wiki/Q-learning



```python
#Hiperparámetros
#Valores del repo
total_episodes = 50000
total_test_episodes = 100
max_steps  = 99

learning_rate = 0.7
gamma = 0.618

#Exploración
epsilon = 1.0
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.01
```

```python
#Algoritmo QLearning para taxi v3


for episode in range(total_episodes):
    
    state = env.reset()
    step = 0
    done = False
    
    for step in range(max_steps):
        # Randomizando
        exp_exp_tradeoff = random.uniform(0,1)
        
   
        if exp_exp_tradeoff > epsilon:
            action = np.argmax(qtable[state,:])
        
      
        else:
            action = env.action_space.sample()
        
        # Take the action (a) and observe the outcome state(s') and reward (r)
        new_state, reward, done, info = env.step(action)

      
        qtable[state, action] = qtable[state, action] + learning_rate * (reward + gamma * 
                                    np.max(qtable[new_state, :]) - qtable[state, action])

        state = new_state
        

        if done == True: 
            break
    
   
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
```

```python
env.reset()
rewards = []

for episode in range(total_test_episodes):
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0

    for step in range(max_steps):
        # Para que se visualice el juego a medida que avanzan los episodes.
        env.render()
        
        action = np.argmax(qtable[state,:])
        
        new_state, reward, done, info = env.step(action)
        
        total_rewards += reward
        
        if done:
            rewards.append(total_rewards)
            #print ("Score", total_rewards)
            break
        state = new_state
env.close()
print ("Score over time: " +  str(sum(rewards)/total_test_episodes))
```

##Exploración y beneficio

El learning rate determina hasta qué punto la información adquirida sobrescribe la información vieja. Un factor de 0 hace que el agente no aprenda (únicamente aprovechando el conocimiento previo), mientras un factor de 1 hace que el agente considere solo la información más reciente (ignorando el conocimiento previo para explorar posibilidades).


##Optimización

Para optimizar el modelo se deben fijar valores de exploración adecuados 

Cuando el agente explora mejora su conocimiento y obtiene mejores recompensas en el largo plazo. 

El learning rate se deriva de estos conceptos y su variación afecta directametne al agente y su aprendizaje. Al modificar este parámetro cambia el tiempo de juego del agente. 

El score no es el mismo cada vez que se ejecuta el programa.
